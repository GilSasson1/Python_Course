# Introduction

## About The Dataset
### Context

Most of us know someone who struggled with breast cancer, or at least heard about the struggles facing patients who are fighting against breast cancer. Breast cancer is the most frequent cancer among women, impacting 2.1 million women each year. Breast cancer causes the greatest number of cancer-related deaths among women. In 2018 alone, it is estimated that 627,000 women died from breast cancer.

The most important part of a process of clinical decision-making in patients with cancers, in general, is the accurate estimation of prognosis and survival duration. Breast cancer patients with the same stage of the disease and the same clinical characteristics can have different treatment responses and overall survival, but why?

Cancers are associated with genetic abnormalities. Gene expression measures the level of gene activity in a tissue and gives information about its complex activities. Comparing the genes expressed in normal and diseased tissue can bring better insights into the cancer prognosis and outcomes. Using machine learning techniques on genetic data has the potentials of giving the correct estimation of survival time and can prevent unnecessary surgical and treatment procedures.

### Content

The Molecular Taxonomy of Breast Cancer International Consortium (METABRIC) database is a Canada-UK Project which contains targeted sequencing data of 1,980 primary breast cancer samples. Clinical and genomic data was downloaded from cBioPortal.

The dataset was collected by Professor Carlos Caldas from Cambridge Research Institute and Professor Sam Aparicio from the British Columbia Cancer Centre in Canada and published on Nature Communications (Pereira et al., 2016). It was also featured in multiple papers including Nature and others:

Associations between genomic stratification of breast cancer and centrally reviewed tumor pathology in the METABRIC cohort
Predicting Outcomes of Hormone and Chemotherapy in the Molecular Taxonomy of Breast Cancer International Consortium (METABRIC) Study by Biochemically-inspired Machine Learning

#### Genetic attributes in the dataset:
The genetics part of the dataset contains m-RNA levels z-score for 331 genes, and mutation for 175 genes.
The formula for Z-score is :

z = (expression in tumor sample - mean expression in reference sample) / standard deviation of expression in reference sample

### Problem Statement
The goal of this project is predicting overall survival (is the patient alive or dead) of breast cancer patients based on genetic and clinical data.

### Feature Selection and Clustering

'death_from_cancer' and 'overall_survival_months' might be proxies for overall_survival this could cause data leakage. Data leakage is when the model is exposed during training to information that it wouldn't have access to in a real-world scenario. This can cause the model to perform better than it should on unseen data.
Lets check the correlation between 'death_from_cancer' and 'overall_survival_months' with 'overall_survival'.

```python
data['death_from_cancer'].replace({'Living': 0, 'Died of Disease': 1, 'Died of Other Causes': 2}, inplace=True)
data[['overall_survival','death_from_cancer', 'overall_survival_months']].corr()
```

```python
# 'death_from_cancer' is obviously problematic, so I will drop it.
data = data.drop('death_from_cancer', axis=1)
y = data['overall_survival']
features = data.drop('overall_survival', axis=1)

# One-hot encoding
features = pd.get_dummies(features, drop_first=True)
# Compute the F-score for each feature and target for feature selection
f_score, p_val = f_classif(features, y)
# Plot the top 50 features with the highest F-score
f_scores = pd.Series(f_score, index=features.columns)
f_scores.sort_values(ascending=False, inplace=True)
f_scores[:50].plot(kind='bar', figsize=(15, 6))
plt.title('Feature with The Highest Association to Target', fontsize=15)
plt.xlabel('Features', fontsize=12)
plt.ylabel('F-Statistic', fontsize=12)
plt.show()
```

```python
# k-means clustering
kmeans = KMeans(n_clusters=2, random_state=73)
kmeans.fit(features)
clusters = kmeans.labels_
print('Silhouette Score:', silhouette_score(features, clusters)) 
sns.scatterplot(x='overall_survival_months', y='age_at_diagnosis', data=features, hue=clusters , palette='Set2')
plt.title('KMeans Clustering', fontsize=15)
plt.xlabel('Time of Survival Since Intervention (Months)', fontsize=12)
plt.ylabel('Age of Diagnosis', fontsize=12)
plt.show()
```

```python
# Chi-Square Test of Independence
# I performed a Chi-Square test to determine if there is a significant association between
# the clusters overall survival.

# Create a contingency table
contingency_table = pd.crosstab(clusters, data['overall_survival'])

# Perform the chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

# Print the results
print(f"Chi2: {chi2}, p-value: {p}, Degrees of Freedom: {dof}")
# plot the contingency table
contingency_table.plot(kind='bar', figsize=(10, 6))
plt.title('Contingency Table', fontsize=15)
plt.xlabel('Cluster', fontsize=12)
plt.ylabel('Overall Survival', fontsize=12)
plt.xticks(ticks=[0, 1], labels=['Cluster 0', 'Cluster 1'])
plt.show()
```

Its seems that 'overall_survival_months' is explaining the clusters well, and also the clusters are significantly associated with the overall survival. Which might indicate about data leakage, so I will drop 'overall_survival_months' too.

```python
print(features.shape[1]) # Since there are a lot of categorical features, one hot encoding has increased the number of features significantly.
# I will use the top 25% of the features with the highest F-score to reduce data dimensionality.
top_features = features[f_scores[:int(features.shape[1] * 0.25 + 1)].index]
top_features = top_features.drop('overall_survival_months', axis=1)
```

# Results and Model Implementation

```python
# Split the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(top_features, y, test_size=0.2, random_state=73)

# Scale the numerical data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train[X_train.select_dtypes(include=[np.float64]).columns])
X_test_scaled = scaler.transform(X_test[X_test.select_dtypes(include=[np.float64]).columns])

# Concatenate the scaled data with the categorical data
X_train_scaled = np.concatenate((X_train_scaled, X_train[X_train.select_dtypes(exclude=[np.float64]).columns]), axis=1)
X_test_scaled = np.concatenate((X_test_scaled, X_test[X_test.select_dtypes(exclude=[np.float64]).columns]), axis=1)
```

```python
# KNN Classifier
knn = KNeighborsClassifier()
knn_cv = cross_val_score(knn, X_train_scaled, y_train, cv=5)
print('Train Accuracy:', round(knn.fit(X_train_scaled, y_train).score(X_train_scaled, y_train), 3))
print('Cross Validation Accuracy:', round(knn_cv.mean(), 3))
```

```python
# Logistic Regression
logistic_regression = LogisticRegression(random_state=73)
cv_scores = cross_val_score(logistic_regression, X_train_scaled, y_train, cv=10)
print('Train Accuracy:', round(logistic_regression.fit(X_train_scaled, y_train).score(X_train_scaled, y_train), 3))
print('Cross Validation Accuracy:', round(cv_scores.mean(), 3))
```

Let's try to improve performance by hyperparameter tuning and using more complex models.

```python
# Decision Tree
dt = DecisionTreeClassifier(random_state=73)
dt_best = GridSearchCV(dt, param_grid={'max_depth': range(1, 10), 'min_samples_leaf': range(1, 10), }, cv=5)
dt_best.fit(X_train_scaled, y_train)
print("Train Accuracy", round(dt_best.score(X_train_scaled, y_train), 3))
print("Best Parameters", dt_best.best_params_)
print("CV Accuracy", round(dt_best.best_score_, 3))
```

```python
# SVM
svm = SVC(probability=True, random_state=73, kernel='rbf')
grid = {'gamma': [1, 0.1, 0.01, 0.001]}
SVM_best = GridSearchCV(svm, grid, cv=5, n_jobs=-1)
SVM_best.fit(X_train_scaled, y_train)
print("Train Accuracy:", round(SVM_best.score(X_train_scaled, y_train), 3))
print('Best Parameters:', SVM_best.best_params_)
print('CV Accuracy:', round(SVM_best.best_score_, 3))
```

```python
# Random Forest
rf = RandomForestClassifier(random_state=73)
params = {'n_estimators': [100, 200, 300, 400, 500], 'max_depth': [1, 5, 1], 'max_features':['sqrt', 'log', None]}
rf_best = GridSearchCV(rf, params, cv=5, n_jobs=-1)
rf_best.fit(X_train_scaled, y_train)
print('Train Accuracy:', round(rf_best.score(X_train_scaled, y_train), 3))
print('Best Parameters:', rf_best.best_params_)
print('CV Accuracy:', round(rf_best.best_score_, 3))
```

```python
# Gradient Boosting
xgb = GradientBoostingClassifier(random_state=73)
param_grid = {'n_estimators': [100, 200, 300, 500], 'max_depth': [1, 5, 7], 'learning_rate': [0.01, 0.1, 1]}
xgb_best = GridSearchCV(xgb, param_grid=param_grid, cv=5, n_jobs=-1,)
xgb_best.fit(X_train_scaled, y_train)
print("Train Accuracy", round(xgb_best.score(X_train_scaled, y_train), 3))
print("Best Parameters", xgb_best.best_params_)
print("CV Accuracy", round(xgb_best.best_score_, 3))
```

### Model Evaluation

```python
# Plot ROC curve of best models on test set
plt.figure(figsize=(10, 6))
models = [knn, logistic_regression, dt_best, SVM_best, rf_best, xgb_best]
models_names =['KNN', 'Logistic Regression', 'Decision Tree','SVM', 'Random Forest', 'Gradient Boosting']
for model, name in zip(models, models_names):
    y_prob = model.predict_proba(X_test_scaled)[:, 1]
    roc_auc = roc_auc_score(y_test, y_prob)
    fpr, tpr, thresholds = roc_curve(y_test, y_prob)
    plt.plot(fpr, tpr, label=f'{name} AUC = {roc_auc:.2f}')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curves of Best Models')
plt.legend(loc='lower right')
plt.show()
```

```python
# Plot confusion matrix
plt.figure(figsize=(8, 15))
for model in models:
    plt.subplot(len(models) // 2 + 1, 2, models.index(model) + 1)
    y_pred = model.predict(X_test_scaled)
    sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap='coolwarm', cbar=False)
    plt.title(models_names[models.index(model)])
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
plt.suptitle('Confusion Matrices of Best Models', fontsize=18, y=0.99)
plt.tight_layout()
plt.show()
```

```python
# Recall Score of best models on test set
for model in models:
    y_pred = model.predict(X_test_scaled)
    print(f'{models_names[models.index(model)]} Recall:', round(recall_score(y_test, y_pred), 3))
```

### Challenges and Possible Improvements 

Challenges:

2. Data leakage: the dataset had 2 columns that were directly related to the target variable, which caused data leakage, this could have resulted in the model to performing better than it should on the test data. Identifying these columns and dropping them was crucial to prevent data leakage.
3. Performance: the dataset had a lot of categorical features which required encoding to be used in the models, this resulted in a drastic increase in the number of features which affected the model's performance and training time. Consequently, I had to reduce the number of features using feature selection techniques. Additionally, utilizing all of CPU's cores while training the models was crucial for reduce training time. This was achieved by setting the parameter 'n_jobs' to -1 in the GridSearchCV function. Essentially bypassing the GIL (Global Interpreter Lock) in Python.

Testing the code frequently and using print statements to debug the code was crucial to identify and resolve these issues.
